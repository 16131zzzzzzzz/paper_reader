```python
from paper_reader.src.pdf_reader import pdf_reader
import os
```

如果你熟悉colab的路径pdf放在哪儿都不要紧，把路径写对就行

如果不熟悉路径此处可以点开左侧边栏第四个文件夹图标，找到paper_reader中的resources文件夹，把自己的pdf放进去，然后pdf_file里面pdf的名称改成你的pdf的名称。

记得填自己的openai_api_key，可以在这里找到自己的openai_key https://platform.openai.com/account/api-keys

```python
pdf_file = "paper_reader/resources/pdfs/1998.pdf" #@param {type:"string"}
output_dir = "drive/MyDrive/output" #@param {type: "string"}
openai_api_key = "" #@param {type: "string"}
if not os.path.exists(pdf_file):
  raise FileNotFoundError(f"{pdf_file} does not exist")
```

```python
pr = pdf_reader(pdf_file, output_dir, "lcnet", "CPU", openai_api_key)
```

如果之前看过这篇文章，可以点上if_jump_ocr跳过ocr的过程（pdf名字需要填之前的），一般不点上。

```python
if_jump_ocr = False #@param {type:"boolean"}
if not if_jump_ocr:
  pr.pdf_to_txt()
pr.chatgpt_embeding(readTextFile=True)
```

```python
question = "I want make ppt to report this paper, list the title and content of the each page" #@param {type: "string"}
pr.ask(question)
```
